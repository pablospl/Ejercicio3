```python
from faker import Faker
import pandas as pd
fake = Faker()
import csv
from random import randint
import numpy as np
import matplotlib.pyplot as plt
productos=pd.read_csv("comercios.csv")

def funcionMostrarProductosRespectoZona(comunidad):
    catalogo=['agua','trigo','huevos','aceite']
    for x in catalogo:
        productosComunidad=(productos.loc[productos['zonas']==comunidad].loc[productos['productos']==x])
      
      
        grafica=productosComunidad[['precio','productos']]
        graficaProductos=grafica['productos']
        graficaPrecios=grafica['precio']
        plt.pie(graficaPrecios,labels=graficaProductos)
        plt.axis=('equal')
        plt.tight_layout()
        plt.legend(graficaProductos,title="productos",loc="center right")
        plt.title("gato")
        plt.show()

seleccionar=str(input("Escriba la comunidad Aut√≥notoma: "))
funcionMostrarProductosRespectoZona(seleccionar)

```
